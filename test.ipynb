```python
def gaussian(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma)**2) / (sigma * np.sqrt(2 * np.pi))

x = np.linspace(-5, 5, 100)
plt.plot(x, gaussian(x, -3, 1) + gaussian(x, 3, 1))
plt.show()
```

```python
def sample_gmm(n_samples, weights, mus, sigmas):
    # Ensure weights sum to 1
    weights = weights / np.sum(weights)
    
    # Choose components based on weights for all samples
    components = np.random.choice(len(weights), size=n_samples, p=weights)
    
    # Generate samples from the chosen components
    samples = np.array([np.random.normal(loc=mus[comp], scale=sigmas[comp]) for comp in components])
    
    return samples

weights = np.array([0.5, 0.5])
mus = np.array([-3, 3])
sigmas = np.array([1, 1])

samples = sample_gmm(100000, weights, mus, sigmas)
plt.hist(samples, bins=100)
plt.show()
```
